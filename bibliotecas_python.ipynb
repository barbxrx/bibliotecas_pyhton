## PANDAS

```python
import pandas as pd
```

```python
retail = pd.read_csv('Train.csv', sep = ';')
```

```python
#Exibe as primeiras linhas
retail.head()
```

```python
#Exibe as últimas linhas
retail.tail()
```

```python
retail.info()
```

```python
#Exibir os campos nulos em ligações feitas
retail[retail["ligacoes_feitas"].isnull
```

```python
#Cria uma nova tabela e elimina os nulos da tabela
new_retail = retail.dropna()
new_retail.info()
```

```python
#Remove valores nulos do dataframe original
retail.dropna(inplace = True)
retail.info()
```

```python
#Substitui os nulos por um valor específico
retail.fillna(4, inplace = True)
```

```python
#Exibir os campos nulos em ligações feitas
retail[retail["ID"]==649]
```

```python
#Substitui os nulos de uma determinada coluna
retail["ligacoes_feitas"].fillna(3, inplace = True)
```

```python
#Exibir que eram nulos e foram substituidos
retail[retail["ID"]==649]
```

```python
#Substituir valores pela média da coluna

media_ligacoes = retail["ligacoes_feitas"].mean()

retail["ligacoes_feitas"].fillna(media_ligacoes, inplace = True)
```

```python
#Exibir que eram nulos e foram substituidos
retail[retail["ID"]==649]
```

```python
#Substituir valores pela mediana da coluna

mediana_ligacoes = retail["ligacoes_feitas"].median()
mediana_desconto = retail["desconto"].median()

retail["ligacoes_feitas"].fillna(mediana_ligacoes, inplace = True)
retail["desconto"].fillna(mediana_desconto, inplace = True)
```

```python
#Substituir valores pela moda da coluna

moda_ligacoes = retail["ligacoes_feitas"].mode()

retail["ligacoes_feitas"].fillna(moda_ligacoes, inplace = True)
```

```python
#Transforma as datas para o tipo datetime
retail['data_envio'] = pd.to_datetime(retail['data_envio'])
retail['data_chegada'] = pd.to_datetime(retail['data_chegada'])
retail.info()
```

```python
retail.head()
```

```python
#Alguns filtros - E
retail[(retail["modo_transporte"]=="Flight") & (retail["ranking_cliente"]==2)]
```

```python
#Alguns filtros - OU
retail[(retail["modo_transporte"]=="Flight") | (retail["modo_transporte"]=="Ship")]
```

```python
#Removendo duplicadas
retail.drop_duplicates(inplace = True)
```

```python
#Selecionar determinadas colunas 
var_num = retail[["ligacoes_feitas","custo_produto","compra_previa","desconto","peso_em_gramas"]]
var_num.head()
```

## Funções de agregação

```python
#Agrupar colunas
moda_ligacoes = retail["ligacoes_feitas"].mean()
moda_ligacoes
```

```python
#Agrupar por uma coluna e dar a média das demais
media_ligacoes = retail.groupby(['ligacoes_feitas']).agg('mean').reset_index()
media_ligacoes
```

```python
#Agrupar uma coluna
agg_ligacoes_preco = retail.groupby(['ligacoes_feitas'])['custo_produto'].agg('mean').reset_index()
agg_ligacoes_preco
```

```python
#Agrupar uma coluna
agg_data_preco = retail.groupby(['data_chegada'])['custo_produto'].agg('mean').reset_index()
agg_data_preco
```

```python
#Agrupar mais de uma coluna
agg_ligacoes_rank_preco = retail.groupby(['ligacoes_feitas','ranking_cliente'])['custo_produto'].agg('mean').reset_index()
agg_ligacoes_rank_preco
```

```python
#Agrupar a mesma coluna por diversas métricas

grouped_single = retail.groupby('ligacoes_feitas').agg({'custo_produto': ['mean', 'min', 'max']})
grouped_single
```

```python
#Agrupar mais de uma coluna

retail.groupby('ligacoes_feitas').agg(
             custo_media = ('custo_produto', 'mean'),
             peso_media = ('peso_em_gramas', 'mean'))
```

## Operações com array

```python
#Adição
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.add(arr1, arr2)

print(newarr)
```

```python
#Subtração
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.subtract(arr1, arr2)

print(newarr)
```

```python
#Multiplicação
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.multiply(arr1, arr2)

print(newarr)
```

```python
#Divisão
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 10, 8, 2, 33])

newarr = np.divide(arr1, arr2)

print(newarr)
```
